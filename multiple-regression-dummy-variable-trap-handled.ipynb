# 1. Introduction To Linear Assumption

Linear regression is used when the dataset has a linear correlation. Before building a linear regression model assumptions are to be validated. If the assumptions are violated, different methodologies must be used.Simple linear regression has one independent variable (predictor) and a dependent variable(response) and multiple linear regression has more than one predictor to predict response.

The simple linear regression equation is represented as

Y = Œ≤0 + Œ≤1 (X1 )+Œµ

Multiple linear regression equation is represented as

Y = Œ≤0 + Œ≤1 (X1 )+Œ≤2 (X 2)+Œ≤3(X3)+Œ≤4(X4)+Œµ

# 2. Different Linear Assumptions

Assumptions of Linear Regression Analysis are :

* Linearity : 
    For linear regression analysis, there must be a linear relationship between predictor(s) and response. Linear relations can be visualized by plotting data points using a scatter plot.
    
* No Heteroskedasticity :
    For a linear relationship if the variance is not a constant and it increases as the predictor (X) increases, then it is called heteroskedasticity. In such cases, the standard error in the output cannot be relied on but still, coefficients will be unbiased.
    
* No omitted variable bias :
    The predictor(s) shouldn‚Äôt be correlated with the error term. It can be represented below equation.
    
    ùúéùëãùúÄ= 0: ‚àÄùë•,ùúÄ

    If predictor(s) are correlated it is called omitted variable bias. It happens when a predictor is excluded and it gets reflected in the error term which leads to biased and counterintuitive estimates. It can be detected by checking the correlation between predictors.
    
* Normality of error :
    We assume error Œµ, is normally distributed, i.e, the mean is equal to zero ( the sum of all errors will be equal to zero (0) or almost 0) and variance ùúé2 of error terms are equal.
    
    Œµ~ N(0,ùúé2)
    
    Normality is violated when the variance of the errors is not consistent across observations.
    
* No autocorrelation :
    Error term values should not have any identifiable relationship.
    
    ùúéùúÄùëñùúÄùëó= 0: ‚àÄùëñ‚â†ùëó
    
    If there exists any relationship between values of error term then autocorrelation comes into the picture. Autocorrelation is also known as serial correlation which affects the standard errors without affecting coefficients that will be unbiased. 

* No multicollinearity :
   Multicollinearity occurs when predictors are themselves correlated which leads to Dummy Variable Trap. To find if multicollinearity exists between predictors correlation between all pairs of predictors is to be checked. In case the correlation coefficient is very high it clearly indicates the predictor is explained by another predictor.

# 4. Dummy Variable Trap

Dummy Variable Trap occurs when predictors are multicollinear, that is one predictor can predict another.

For this dataset, fourth column D contains categorical data. Before building a model categorical data is to be encoded.For encoding column D categorical data, lets us take 3 Dummy Variable 
1. D1 for State_California
1. D2 for State_Florida 
1. D3 for State_New York

Let the Matrix representation of State_California be [100] and State_Florida be [010] and State_New York be [001].

D1, D2 and D3 are highly correlated or multicollinear because if D1 and D2 are zero it is obvious D3 will be 1. ie., if the data does not belong to State_California or State_Florida then it is sure, data belongs to State_NewYork. This means here two dummy variables can predict the third one.

Or in a broader sense, it can be said, if one dummy variable is dropped no relevant information will be lost as other dummy variables can explain dropped feature fully. So if there are n dummy variables, n-1 dummy variables will be able to predict the value of the n-th dummy variable. So in order to avoid the dummy variable trap, one dummy variable has to be dropped.

# 5. Mathematical Analysis of Dummy Variable Trap

Multiple linear regression equation is represented as

Y = Œ≤0 + Œ≤1 (X1 )+Œ≤2 (X 2)+Œ≤3(X3)+Œ≤4(X4)+Œµ

After assigning dummy variable, regression model can be represented as

Y = Œ≤0 + Œ≤1 (X1 )+Œ≤2 (X 2)+Œ≤3(X3)+Œ≤4D1+Œ≤5D2+Œ≤6D3+Œµ ‚Äî equation 1


we have D1+D2+D3 = 1
or D3 = 1- (D1+D2) ‚Äî equation 2 which indicates D3 is perfectly explained by the other two dummy variables D1 and D2


If we use equation 2 in 1, we get

Y = Œ≤0 + Œ≤1 (X1 )+Œ≤2 (X 2)+Œ≤3(X3)+Œ≤4D1+Œ≤5D2+Œ≤6(1‚Äì(D1+D2))+Œµ

Y = Œ≤0 + Œ≤1 (X1 )+Œ≤2 (X 2)+Œ≤3(X3)+Œ≤4D1+Œ≤5D2+Œ≤6-Œ≤6D1-Œ≤6D2+Œµ

Y =( Œ≤0+ Œ≤6)+ Œ≤1 (X1 )+Œ≤2 (X 2)+Œ≤3(X3)+(Œ≤4-Œ≤6)D1+(Œ≤5 -Œ≤6)D2+Œµ

Y = Œ≤0*+ Œ≤1 (X1 )+Œ≤2 (X 2)+Œ≤3(X3)+Œ≤4*D1+Œ≤5*D2 +Œµ ‚Äî equation 3


From the equation 3, it is clear that coefficient Œ≤0, Œ≤4 and Œ≤5 is impacted by D3 coefficient Œ≤6. So one dummy variable should be dropped to avoid trapped by dummy variable.

As the fourth column, ‚Äòstate‚Äô contains categorical data, it should be encoded before building the model.

For enoding categorical data, if we take 3 Dummy Variable D1 for State_California, D2 for State_Florida and D3 for State_New York it will lead to a Dummy variable trap. 

Dummy Variables D1, D2 and D3 are highly correlated or multicollinear because if D1 and D2 are zero it is obvious D3 is 1. ie., if the data does not belong to State_California or State_Florida then it is sure, the data belongs to State_NewYork.

So to avoid the dummy variable trap we have to drop one dummy variable while building the model.

That can be done by X=pd.get_dummies(X, drop_first=True), which will drop one dummy variable.

```python
X=pd.get_dummies(X, drop_first=True)
```

```python
X
```

Here ‚Äòdrop_first = True‚Äô have dropped dummy variable for State_California.

```python
#Splitting testdata into X_train,X_test,y_train,y_test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.33,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
#Prediciting Y from Linear regression Model
y_pred = lr.predict(X_test)
```

```python
print(y_pred)
```

From below difference between predicted and actual y can be seen

```python
df=pd.DataFrame(data={'Predicted value':y_pred,'Actual Value':y_test})
df
```

```python
from sklearn import metrics
print('Mean Absolute Error      :', metrics.mean_absolute_error(y_test , y_pred))
print('Mean Squared Error       :', metrics.mean_squared_error(y_test , y_pred))
print('Root Mean Squared Error  :', np.sqrt(metrics.mean_squared_error(y_test , y_pred)))
print('R Square                 :', metrics.r2_score(y_test , y_pred))
```
